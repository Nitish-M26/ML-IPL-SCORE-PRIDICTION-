```python
!sudo apt install tesseract-ocr
!pip install pytesseract opencv-python-headless

import cv2
import pytesseract
import numpy as np

# Load the image
uploaded = files.upload()
image_file = next(iter(uploaded.values()))
img = cv2.imdecode(np.frombuffer(image_file, np.uint8), cv2.IMREAD_UNCHANGED)

# Preprocessing
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

# Perform OCR using pytesseract
text = pytesseract.image_to_string(thresh, config='--psm 6')

# Print the extracted text
print(text)
```

**For other language like Latin**

```python
!sudo apt install tesseract-ocr-lat # install the Latin language data for Tesseract
!pip install googletrans==3.1.0a0
from googletrans import Translator

# Load the image
uploaded = files.upload()
image_file = next(iter(uploaded.values()))
img = cv2.imdecode(np.frombuffer(image_file, np.uint8), cv2.IMREAD_UNCHANGED)

# Preprocessing (adjust if necessary)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

# Perform OCR
latin_text = pytesseract.image_to_string(thresh, config='--psm 6 -l lat') # use -l lat to specify Latin language

# Translate to English
translator = Translator()
translation = translator.translate(latin_text, dest='en')

# Print results
print("Latin Text:")
print(latin_text)
print("\nEnglish Translation:")
print(translation.text)
```

**For blurred Image**

```python
import cv2
import pytesseract #removed extra space at the beginning of the line
import numpy as np
from google.colab import files
ploaded = files.upload()
image_file = next(iter(uploaded.values()))
img = cv2.imdecode(np.frombuffer(image_file, np.uint8), cv2.IMREAD_UNCHANGED)
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

# Denoise using Gaussian Blur
blur = cv2.GaussianBlur(thresh, (5,5), 0)
# Perform OCR
text = pytesseract.image_to_string(blur, config='--psm 6')

# Print the extracted text
print(text)
```
